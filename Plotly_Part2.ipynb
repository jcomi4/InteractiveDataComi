```python
import pandas as pd
import plotly.graph_objects as go
```

```python
meteor = pd.read_csv('Meteorite_Landings_Sample(1).csv')
meteor.head()
```

```python
meteor.index
```

```python
fig1 = go.Figure(data=go.Scattergeo(
    lon = meteor['reclong'],
    lat = meteor['reclat'],
    mode='markers',
))

fig1.update_layout(
    title = 'Meteorite Landings',
    geo_scope = 'world',
)

fig1.show()
```

```python
fig2 = go.Figure(data=go.Scattergeo(
    lon = meteor['reclong'],
    lat = meteor['reclat'],
    mode='markers',
    marker = dict(
        size = 8,
        opacity = 0.8,
        symbol = 'star-diamond',
        color = 'red'
    )
))

fig2.update_geos(projection_type = 'equirectangular')

fig2.update_layout(
    title = 'Meteorite Landings',
    geo_scope = 'world',
)

fig2.show()
```

```python
fig3 = go.Figure(data=go.Scattergeo(
    lon = meteor['reclong'],
    lat = meteor['reclat'],
    mode='markers',
    marker = dict(
        size = 8,
        opacity = 0.8,
        reversescale = True,
        autocolorscale = False,
        symbol = 'circle',
        colorscale = 'electric',
        color = meteor['mass (g)'],
        colorbar = dict(
            title = dict(
                text = 'Meteorite Mass (g)'
            )
        )
    )
))

fig3.update_geos(projection_type = 'equirectangular',
                showland = True, landcolor = 'green',
                showocean = True, oceancolor = 'lightblue',
                showlakes = True, lakecolor = 'grey',
                showrivers = True, rivercolor = 'grey')

fig3.update_layout(
    title = 'Meteorite Landings',
    geo_scope = 'world',
)

fig3.show()
```

```python
fig4 = go.Figure(data=go.Scattergeo(
    lon = meteor['reclong'],
    lat = meteor['reclat'],
    mode='markers',
    marker = dict(
        size = 8,
        opacity = 0.8,
        reversescale = True,
        autocolorscale = False,
        symbol = 'circle',
        colorscale = 'electric',
        color = meteor['mass (g)'],
        colorbar = dict(
            title = dict(
                text = 'Meteorite Mass (g)'
            )
        )
    )
))

fig4.update_geos(projection_type = 'orthographic',
                showland = True, landcolor = 'green',
                showocean = True, oceancolor = 'lightblue',
                showlakes = True, lakecolor = 'grey',
                showrivers = True, rivercolor = 'grey')

fig4.update_layout(
    title = 'Meteorite Landings',
    geo_scope = 'world',
)

fig4.show()
```

```python
fig5 = go.Figure()

fig5.add_trace(
    go.Scattergeo(
        lon = meteor['reclong'],
    lat = meteor['reclat'],
    mode='markers',
    marker = dict(
        size = 8,
        opacity = 0.8,
        reversescale = True,
        autocolorscale = False,
        symbol = 'circle',
        colorscale = 'electric',
        color = meteor['mass (g)'],
        colorbar = dict(
            title = dict(
                text = 'Meteorite Mass (g)'
            )
        )
    )
    )
)

fig5.update_geos(projection_type = 'equirectangular',
                showland = True, landcolor = 'green',
                showocean = True, oceancolor = 'lightblue',
                showlakes = True, lakecolor = 'grey',
                showrivers = True, rivercolor = 'grey')


button1 = {
    'args': [{'geo':{'projection':{'type': 'equirectangular'}}}],
    'label': 'Map', 
    'method': 'relayout'
}

button2 = {
    'args': [{'geo':{'projection':{'type': 'orthographic'}}}],
    'label': 'Globe', 
    'method': 'relayout'
}

fig5.update_layout(
    title = 'Meteorite Landings',
    geo_scope = 'world',
    updatemenus = [
        dict(
            buttons = [button1, button2],
            direction = 'down',
            x = 0.1,
            xanchor = 'left',
            y = 1.1,
            yanchor = 'top',
        )
    ]
)

fig5.show()
```

```python
meteor['recclass'].unique
```

```python
recclass_list = list(meteor['recclass'].unique())
```

```python
meteor[meteor['recclass'] == 'L5']['mass']
```

```python
falses = [False] * len(recclass_list)
recclass_buttons = []
fig8 = go.Figure()

for i in range(0, len(recclass_list)): 
    fig8.add_traces(go.Box(y = meteor[meteor['recclass'] == str(recclass_list[i])]['mass'], name = str(recclass_list[i])))
    new_falses = falses.copy()
    new_falses.insert(i, True)
    recclass_buttons.append({'label':str(recclass_list[i]),
                            'method': 'restyle',
                            'args': [{'visible':new_falses}]})

show_all_button = {'label':'Show All',
                            'method': 'restyle',
                            'args': [{'visible': [True] * len(recclass_list)}]}

recclass_buttons.insert(0, show_all_button)


fig8.update_layout(updatemenus = [{'type': 'dropdown', 
                                  'direction': 'right', 
                                  'buttons': recclass_buttons}])

fig8.show()
```
